```python
def ecode_units(x):
    if x=='0':
        return 0
    if x=='1':
        return 1

Basket_set=df1.applymap(ecode_units)
Basket_set.info()
```

# Extracting rules

```python
frequent_itemset=apriori(Basket_set,min_support=0.1,use_colnames=True)
```

```python
frequent_itemset
```

```python
rules=association_rules(frequent_itemset,metric='lift',min_threshold=1)
rules
```

```python
sorted_rules=rules.sort_values(['lift'],ascending=False)
sorted_rules.head(10)
```
